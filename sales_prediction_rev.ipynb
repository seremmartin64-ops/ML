  READING THE DATASET

```python
import kagglehub
import pandas as pd
import os
```

```python
# Download latest version
path = kagglehub.dataset_download("bumba5341/advertisingcsv")
print(path)
print(os.listdir(path))

file_path = os.path.join(path, "Advertising.csv")
data = pd.read_csv(file_path)
data
```

DATA CLEANING

```python
# Ensures you are working with clean dataset
# In the above dataset we have un nessesary called the Unnamed: 0
data.drop(columns = ["Unnamed: 0"], inplace=True)
data
```

SPLITING DATA INTO FEATURES AND OUTPUT

```python
# features = x
# output = y
array = data.values
array.shape
X = array[:,0:3]
Y = array[:,3]
```

SPLITING DATA INTO TRAINING AND TESTING

```python
# Splitting data into training 70% and testing 30%
# Underfitting
# Overfitting
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state = 42)
print(X_train.shape)
print(X_test.shape)
```

HOW TO TRAIN A ML MODEL

```python
# Cross_validation- technique used in ml to asses how a model performs in a new unseen data
# 1).Decision tree
# 2).kneighbours
# 3).linear Regression
# 4).support vector machine
# 5).Random forest
# 6).Gradient boosting
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(X_train, Y_train)
```

EVALUATION

```python
# Evaluation is the process of determining the accuracy of ML model
# Regression:r2 score:MeanAbsoluteError
Model_prediction = model.predict(X_test)
print(Model_prediction)

print("=================")
print(Y_test)

# regression- we check the distance between the model and the Y_testresult(actual answer)
# r2 score :MeanAbsoluteerror
print("===================")
from sklearn.metrics import r2_score
accuracy = r2_score(Y_test, Model_prediction)*100
print(accuracy)
```

ACTUAL PREDICTION

```python
# Actual_prediction- we apply ML model to predict our sales
# Expense for adverts
adverts_expenditure = [50,50,40]
future_sales = model.predict([adverts_expenditure])
print(future_sales)
```
